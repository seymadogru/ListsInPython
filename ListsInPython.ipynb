Aşağıdaki listede öğrencilerin aldığı notlar yer almaktadır. En yüksek not kaçtır ve kaç kişi bu notu almıştır?

```python
scores = [99, 87, 78, 86, 68, 94, 76, 97 ,56, 98, 76, 87, 79, 90, 73, 93, 82, 69, 97, 98]
num = scores[0]
sayac=0
for i in scores:
  if i>num:
    num = i
print("En yüksek not:", num )  
for n in scores:
  if num==n:
    sayac+=1
print("Bu notu alan kişi sayısı:", sayac)
```
